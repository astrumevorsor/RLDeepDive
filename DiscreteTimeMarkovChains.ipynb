#Defining a state transition matrix for markov processes



```python

# State Transition Matrix for The Great Weather Conspiracy

# States:
# 0: Sunny
# 1: Cloudy
# 2: Rainy

transition_matrix = [
    [0.7, 0.2, 0.1],  # From Sunny (State 0)
    [0.3, 0.35, 0.35], # From Cloudy (State 1)
    [0.2, 0.3, 0.5]   # From Rainy (State 2)
]

# Explanation:
# Each row represents the current state, and each column represents the probability of transitioning to another state.
# For example, the first row [0.7, 0.2, 0.1] means:
# - From Sunny, there's a 70% chance of staying Sunny.
# - From Sunny, there's a 20% chance of transitioning to Cloudy.
# - From Sunny, there's a 10% chance of transitioning to Rainy.

# You can use this matrix in your Python code to simulate the weather patterns
# over the next 100 days and analyze the probabilities of different weather sequences.
```

#how do you find the state at t+1 given that t was given?

```python

import numpy as np

# Current state (e.g., at time t)
current_state = 0  # Assuming it's Sunny (State 0)

# Convert the transition matrix to a NumPy array for easier calculations
transition_matrix_np = np.array(transition_matrix)

# Find the probabilities of transitioning to each state at t+1
next_state_probabilities = transition_matrix_np[current_state]

# To find the actual state at t+1, you can either:
# 1. Sample from the probability distribution (more realistic simulation):
next_state = np.random.choice([0, 1, 2], p=next_state_probabilities)
print(f"At time t+1, the state is: {next_state}")

# 2. Find the state with the highest probability (simpler but less accurate):
# next_state = np.argmax(next_state_probabilities)
# print(f"At time t+1, the state is: {next_state}")

# Example:
# If the current state is Sunny (state 0), the next_state_probabilities will be [0.7, 0.2, 0.1].
# If we randomly sample from this distribution, we might get:
# - State 0 (Sunny) with a 70% chance
# - State 1 (Cloudy) with a 20% chance
# - State 2 (Rainy) with a 10% chance
```


## Initial State Distribution

The initial state (X₀) can be defined in two ways:
- As a specific deterministic state
- As a probability distribution across the state space[1][2]

## Mathematical Representation

The initial state is typically represented as:
- A row vector μ₀ that specifies the probability distribution over all possible states[2]
- Each element μ₀(x) represents the probability P{X₀ = x} for each state x in the state space Ω

## Properties

The initial distribution has these characteristics:
- Must be a valid probability distribution (sums to 1)
- Can be used with the transition matrix P to determine future distributions:
$$ μₜ = μ₀P^t $$
where μₜ represents the state distribution at time t[2]

## Practical Implementation

When implementing Markov chains, you can:
- Choose a specific starting state (deterministic case)
- Sample from an initial probability distribution (stochastic case)
- Use different initial distributions with the same transition matrix to analyze different scenarios[2]

Therefore, the initial state can be either deterministic or probabilistic, depending on how you define the initial distribution μ₀.


```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set styling for plots
#plt.style.use('seaborn')
sns.set_palette("husl")

# Define transition matrix for a simple weather model
# States: 0 = Sunny, 1 = Cloudy, 2 = Rainy
P = np.array([
    [0.7, 0.2, 0.1],  # Sunny transitions
    [0.3, 0.4, 0.3],  # Cloudy transitions
    [0.2, 0.3, 0.5]   # Rainy transitions
])

# Initial state distribution
initial_state = np.array([1, 0, 0])  # Starting from Sunny state

#We chose this, it needn't necessarily be this as the theory says

def simulate_markov_chain(P, initial_state, n_steps):
    """
    Simulate a Markov chain for n discrete time steps.

    Arguments-

    p: Transition matrix- defines the probabilties of moving between different states in the markov chain

    initial state: numpy array representing the initial probabilities of being in each state

    n_steps: The discrete timesteps you want to simulate the markov chain for

    """
    states = np.zeros(n_steps, dtype=int) #initialize your state array as an array of zeroes, this is a default as in mathematics we'd assume the event as false unless it happened

    states[0] = np.random.choice(len(initial_state), p=initial_state) #This is the initial state of the simulation, we choose an initial state based on the probabilties defined in initial_state
    #The above line of code should output an array of initial states based on the current state and the probabilties

    #print(states)

    #The purpose of the following for loop is to basically populate the state array as per the time steps
    #The probability distribution is still a finite set/discrete array here.
    #The key assumption here is that the t-th state only relies on your (t-1)th state

    for t in range(1, n_steps):
        states[t] = np.random.choice(len(P), p=P[states[t-1]])


    #print (states)

    return states

# Simulate the chain
n_steps = 100
states = simulate_markov_chain(P, initial_state, n_steps)

# Plot the simulation
plt.figure(figsize=(12, 6))
plt.plot(states, marker='o', linestyle='-', markersize=4)
plt.yticks([0, 1, 2], ['Sunny', 'Cloudy', 'Rainy'])
plt.xlabel('Time Step')
plt.ylabel('Weather State')
plt.title('Weather State Simulation using Markov Chain')
plt.grid(True)
plt.show()

# Calculate state distribution over time
def calculate_distribution_over_time(P, initial_state, n_steps):
    """Calculate state distribution over time."""
    distribution = np.zeros((n_steps, len(P)))
    current_dist = initial_state

    for t in range(n_steps):
        distribution[t] = current_dist
        current_dist = np.dot(current_dist, P)

    return distribution

# Calculate and plot state distributions
n_steps = 20
distributions = calculate_distribution_over_time(P, initial_state, n_steps)

plt.figure(figsize=(12, 6))
for i, state in enumerate(['Sunny', 'Cloudy', 'Rainy']):
    plt.plot(distributions[:, i], label=state, marker='o')
plt.xlabel('Time Step')
plt.ylabel('Probability')
plt.title('State Probabilities Over Time')
plt.legend()
plt.grid(True)
plt.show()

# Calculate steady state
def find_steady_state(P, tol=1e-8, max_iter=1000):
    """Find steady state distribution using power iteration."""
    n = len(P)
    pi = np.ones(n) / n

    for _ in range(max_iter):
        pi_next = np.dot(pi, P)
        if np.all(np.abs(pi_next - pi) < tol):
            return pi_next
        pi = pi_next

    return pi

# Print steady state distribution
steady_state = find_steady_state(P)
print("\nSteady State Distribution:")
for i, state in enumerate(['Sunny', 'Cloudy', 'Rainy']):
    print(f"{state}: {steady_state[i]:.3f}")
```
